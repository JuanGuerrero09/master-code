```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# step size
h = 0.1

def compute_solution(h):
    # define grid
    x = np.arange(0, 2*np.pi, h) 
    print(x)
    # compute function
    y = np.cos(x) 
    # compute vector of forward differences
    print(np.diff(y))
    forward_diff = np.diff(y)/h 
    # compute corresponding grid
    x_diff = x[:-1:] 
    # compute exact solution
    exact_solution = -np.sin(x_diff) 

    # Plot solution
    plt.figure(figsize = (6, 4))
    plt.plot(x_diff, forward_diff, '--', \
            label = 'Finite difference approximation')
    plt.plot(x_diff, exact_solution, \
            label = 'Exact solution')
    plt.legend()
    plt.show()

    # Compute max error between 
    # numerical derivative and exact solution
    max_error = max(abs(exact_solution - forward_diff))
    print(max_error)


compute_solution(h)
compute_solution(0.5)
compute_solution(0.01)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros fsicos y numricos
L = 1.0        # Longitud del dominio [m]
nx = 10        # Nmero de celdas
dx = L / (nx-1)
D = 0.1        # Coeficiente de difusin
dt = 0.01      # Paso temporal
nt = 50        # Nmero de pasos de tiempo

# Condiciones iniciales y de frontera
phi = np.zeros(nx)
phi[nx//2] = 1.0   # Pico en el centro
phi_new = phi.copy()

# ---- FTCS (Explcito) ----
r = D * dt / dx**2
phi_FTCS = [phi.copy()]

for n in range(nt):
    for i in range(1, nx-1):
        phi_new[i] = phi[i] + r * (phi[i+1] - 2*phi[i] + phi[i-1])
    phi_new[0], phi_new[-1] = 0, 0   # Dirichlet
    phi[:] = phi_new
    phi_FTCS.append(phi.copy())

# ---- BTCS (Implcito) ----
A = np.zeros((nx, nx))
for i in range(nx):
    if i == 0 or i == nx-1:
        A[i, i] = 1  # Dirichlet BC
    else:
        A[i, i-1] = -r
        A[i, i] = 1 + 2*r
        A[i, i+1] = -r

phi = np.zeros(nx)
phi[nx//2] = 1.0
phi_BTCS = [phi.copy()]

for n in range(nt):
    b = phi.copy()
    phi_new = np.linalg.solve(A, b)
    phi[:] = phi_new
    phi_BTCS.append(phi.copy())

# ---- Visualizacin ----
x = np.linspace(0, L, nx)
plt.figure(figsize=(8, 5))
plt.plot(x, phi_FTCS[0], 'k--', label='Initial')
plt.plot(x, phi_FTCS[20], 'r-', label='FTCS (t=20dt)')
plt.plot(x, phi_BTCS[20], 'b-', label='BTCS (t=20dt)')
plt.xlabel('x')
plt.ylabel('')
plt.legend()
plt.title('Comparison: FTCS vs BTCS for Diffusion (10 cells)')
plt.grid(True)
plt.show()
```
